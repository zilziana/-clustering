```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
data = pd.read_csv('set.csv')

xx = tuple(data['x'])
yy = tuple(data['y'])
kls = tuple(data['kelas'])

dataset = np.array([xx,yy]),np.array(kls)
```

```python
points =  dataset[0]
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
dendrogram = sch.dendrogram(sch.linkage(points, method = 'ward'))
```

```python
plt.scatter(dataset[0][:,0], dataset[0][:,2])
```

```python
hc = AgglomerativeClustering(n_clusters =2, affinity = 'euclidean', linkage = 'ward')
```

```python
y_hc = hc.fit_predict(points)
```

```python
plt.scatter(points[y_hc == 0,0], points[y_hc == 0,1], s = 300, c = 'blue')
plt.scatter(points[y_hc == 1,0], points[y_hc == 1,1], s = 300, c = 'green')

plt.scatter
```
